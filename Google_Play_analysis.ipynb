### Now let's start cleaning the each row

checking any duplicate values in the App Name

```python
df['Installs'].unique() #displaying the unique values
```

```python
df['Installs']=df['Installs'].str.split('+').str[0] #removing the + sign
df['Installs'].replace(',','', regex=True, inplace=True) #removing the comma
df['Installs']=df['Installs'].astype('int64') #converting the datatype to int64
```

```python
df['Installs'].unique() #displaying the unique values
```

```python
df['Currency'].unique() #displaying the unique values
```

```python
df['Size'].unique() #displaying the unique values
```

```python
df['Size']=df['Size'].apply(lambda x: x.replace('M','') if 'M' in str(x) else x) #removing the M
```

```python
df['Size']=df['Size'].apply(lambda x: x.replace('k','') if 'k' in str(x) else x) #removing the k
```

```python
df['Size']=df['Size'].apply(lambda x: x.replace(',','') if 'k' in str(x) else x) #removing the comma
```

```python
df['Size']=df['Size'].apply(lambda x: x.replace('Varies with device','0') if 'Varies with device' in str(x) else x) #removing the Varies with device
```

```python
df['Type']=np.where(df['Free']==True, 'Free', 'Paid') #creating a new column Type
df.drop('Free', axis=1, inplace=True) #dropping the Free column
```

```python
df['Content Rating'].unique() #displaying the unique values
```

```python
# value counts
df['Content Rating'].value_counts()
```

```python
df['Content Rating']=df['Content Rating'].replace('Unrated','Everyone') #replacing the Unrated with Everyone
# everyone 10+ with teen
df['Content Rating']=df['Content Rating'].replace('Everyone 10+','Teen')
#  mature 17+ with adults
df['Content Rating']=df['Content Rating'].replace('Mature 17+','Adults')
# adults 18+ with adults
df['Content Rating']=df['Content Rating'].replace('Adults only 18+','Adults') 
```
